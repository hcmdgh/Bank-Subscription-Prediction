# 1 数据预处理

## 1.1 载入训练集和测试集

```python
import pandas as pd 

train_df = pd.read_csv('./data/train.csv')
test_df = pd.read_csv('./data/test.csv')

train_df
```

```python
test_df 
```

## 1.2 合并训练集和测试集

```python
total_df = pd.concat([train_df, test_df], keys=['train', 'test'])
total_df = total_df.drop(columns='id')
total_df
```

## 1.3 生成onehot特征

```python
def generate_onehot_feature(df: pd.DataFrame,
                            columns: list[str]) -> pd.DataFrame:
    for column in columns:
        df_onehot = pd.get_dummies(df[[column]])
        df = df.join(df_onehot)
        df = df.drop(columns=column)
    
    return df 
```

```python
total_df = generate_onehot_feature(
    total_df, 
    ['contact', 'job', 'marital', 'education', 'default', 'housing', 'loan', 'month', 'day_of_week', 'poutcome'],
)
total_df
```

## 1.4 标签值转为0/1

```python
X = total_df.drop(columns='subscribe')
y = total_df[['subscribe']]

X_train = X.loc['train']
y_train = y.loc['train']
X_test = X.loc['test']
y_test = y.loc['test']
```

```python
y_train['subscribe'] = y_train['subscribe'].map(
    dict(
        no = 0, 
        yes = 1, 
    ),
)
y_train 
```

## 1.5 划分训练集和验证集

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X_train, 
    y_train, 
    test_size = 0.2, 
    random_state = 142,
)

X_train.shape, X_val.shape, y_train.shape, y_val.shape 
```

## 1.6 特征归一化

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)

X_train.min(), X_train.max(), X_train.shape 
```

# 2 训练

```python
import lightgbm as lgb 
from sklearn.model_selection import GridSearchCV

clf = lgb.LGBMClassifier(
    learning_rate = 0.2, 
    metric = 'l1', 
    n_estimators = 30, 
    min_child_samples = 15, 
    num_leaves = 25,
    verbosity = -1,
    max_depth = 8, 
    subsample = 0.9, 
    colsample_bytree = 0.9, 
)

clf.fit(X=X_train, y=y_train)
```

```python
from sklearn.metrics import auc, accuracy_score, roc_auc_score, roc_curve

y_pred = clf.predict(X_val)

val_acc = accuracy_score(y_val, y_pred)

val_acc 
```

# 3 测试

```python
from collections import Counter

y_pred = clf.predict(X_test)

Counter(y_pred.tolist())
```

```python
submission = pd.DataFrame(
    dict(
        id = range(1, len(y_pred) + 1), 
        subscribe = y_pred, 
    )
)

submission['subscribe'] = submission['subscribe'].map({0: 'no', 1: 'yes'})

submission 
```

```python
submission.to_csv('./data/submission.csv', index=False)
```
